```python
import pandas as pd

df = pd.read_spss('../../SPSS-Python/spss-datasets/Mah_60Day_data_practice.sav')

df.head()
```

#### Pull data from Q1

```python
# Specify the variable for which you want to calculate the frequency
variable_name = 'Q1'  # Replace with the name of the variable you're interested in

# Calculate the frequency of responses
variable_counts = df[variable_name].value_counts()
total_responses = variable_counts.sum()
variable_percentages = variable_counts / total_responses * 100

# Display the frequency of responses as percentages
print(f'Frequency of responses for variable "{variable_name}":')
for value, count, percentage in zip(variable_counts.index, variable_counts.values, variable_percentages.values):
    print(f'{value}: {count} ({percentage:.2f}%)')
```

```python
# Specify the variable for which you want to calculate the frequency
variable_name = 'Q1'  # Replace with the name of the variable you're interested in

# Calculate the frequency of responses
variable_counts = df[variable_name].value_counts()
total_responses = variable_counts.sum()
variable_percentages = variable_counts / total_responses * 100

# Sort the results by percentage in descending order
variable_results = pd.concat([variable_counts, variable_percentages], axis=1)
variable_results.columns = ['Frequency', 'Percentage']
variable_results = variable_results.sort_values(by='Percentage', ascending=False)

# Display the frequency of responses as percentages, sorted by percentage
print(f'Frequency of responses for variable "{variable_name}":')
for value, count, percentage in zip(variable_results.index, variable_results['Frequency'], variable_results['Percentage']):
    print(f'{value}: {count} ({percentage:.2f}%)')
```
